```python
numerical_cols = ['num_passengers', 'purchase_lead',
       'length_of_stay', 'flight_hour', 'wants_extra_baggage', 'wants_preferred_seat',
       'wants_in_flight_meals', 'flight_duration', 'booking_complete']
non_numerical_cols = ['sales_channel','trip_type','flight_day','route','booking_origin']
print({i: df[i].unique() for i in non_numerical_cols})
```

```python
# sales channel : one hot encoding
# trip type: one hot encoding
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

one_hot_features = ['sales_channel', 'trip_type']
ct = ColumnTransformer(
    transformers=[('encoder', OneHotEncoder(sparse_output=False), [1,2])], 
    remainder='passthrough'
)

transformed = ct.fit_transform(df)
encoded_feature_names = ct.named_transformers_['encoder'].get_feature_names_out(one_hot_features)

transformed_df = pd.DataFrame(transformed[:,:len(encoded_feature_names)],columns=encoded_feature_names)

df = pd.concat([df["num_passengers"],transformed_df,df.iloc[:,3:]],axis=1)
```

```python
mapping = {
    "Mon": 1,
    "Tue": 2,
    "Wed": 3,
    "Thu": 4,
    "Fri": 5,
    "Sat": 6,
    "Sun": 7,
}

df["flight_day"] = df["flight_day"].map(mapping)
```
